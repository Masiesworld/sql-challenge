```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from config import username,password
import psycopg2
from sqlalchemy import create_engine
%matplotlib inline
```

```python
engine = create_engine(f"postgresql+psycopg2://{username}:{password}@localhost/sql_hw_db")
conn = engine.connect()
```

```python
# Get the table names
engine.table_names()
```

```python
salaries = pd.read_sql("SELECT * FROM Salaries;",conn)
```

```python
# get the salaries dataframe
salaries_df = pd.DataFrame(salaries)
salaries_df.head()
```

```python
# Most Common Employee Salary range
emp_salary_query = '''SELECT e.emp_no,t.title, s.salary 
FROM Employees AS e
JOIN Salaries AS s
ON e.emp_no = s.emp_no
JOIN Titles AS t
ON e.emp_title_id = t.title_id'''
```

```python
emp_salary = pd.read_sql(emp_salary_query,conn)
emp_salary.head(10)
```

```python
# Make the salary_employee freq histogram
plt.figure(figsize=(10,8))
plt.hist(x=emp_salary.salary,bins='auto',color='purple',alpha=0.9,rwidth=2)
plt.xlabel('Salary')
plt.ylabel('Number of Employees')
plt.title("Frequency Distribution of Employee Salary")
plt.grid(axis='y',alpha=0.75)
plt.xlim(38000,110000)
plt.show()

# Most common one is around $40,000.
```

```python
# Average salay by Title
sql_query = '''SELECT t.title, ROUND(AVG(s.salary),2) AS "Average_Salary" 
    FROM Employees AS e
    JOIN Salaries AS s
    ON e.emp_no = s.emp_no
    JOIN Titles AS t
    ON e.emp_title_id = t.title_id
    GROUP BY t.title
    ORDER BY "Average_Salary"'''
```

```python
avg_salary = pd.read_sql(sql_query,conn)
avg_salary
```

```python
title = list(avg_salary['title'])
avg_sa = list(avg_salary['Average_Salary'])

fig,ax = plt.subplots()
width = 0.75
ind = np.arange(len(avg_sa))

ax.barh(ind,avg_sa,width)

ax.set_yticks(ind+width/2)
ax.set_yticklabels(title, minor=False)

plt.xticks(rotation=45)
plt.ylabel('Average Salary in USD($)')
plt.title('Average Salary By Titles')

for i,v in enumerate(avg_sa):
    ax.text(v,i,f"${v}",fontweight = 'bold',va='center')

plt.show()
```
